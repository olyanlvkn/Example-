```python
def check_input(element):
    return isinstance(element, str)


def get_substring(s):
    if not s:
        return ""

    char_index = {}
    start = 0
    max_length = 0
    result = ""

    for i, char in enumerate(s):
        if char in char_index and char_index[char] >= start:
            start = char_index[char] + 1

        char_index[char] = i

        current_length = i - start + 1
        if current_length > max_length:
            max_length = current_length
            result = s[start:i+1]

    return result


def main(lst):

    longest_str = ""
    longest_str_index = -1

    for i, element in enumerate(lst):
        if check_input(element):
            if longest_str_index == -1 or len(element) > len(longest_str):
                longest_str = element
                longest_str_index = i

    if longest_str_index != -1:
        substring = get_substring(longest_str)
    else:
        substring = ""

    print(f"{longest_str_index}, '{substring}'")
```
